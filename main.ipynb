### Challenge 2: Count Commits

Count how many commits were made in the past week.

1. Obtain all the commits made in the past week via API, which is a JSON array that contains multiple commit objects.

1. Count how many commit objects are contained in the array.

````python
response = requests.get("https://api.github.com/repos/ironhack-datalabs/datalis0819/commits")
response = pd.DataFrame(response.json())
response
#len(response)

response['commit'][0]
````

````python

# response['commit'][0]

from pandas.io.json import json_normalize

response = requests.get("https://api.github.com/repos/ironhack-datalabs/datalis0819/commits")
norm_json = json_normalize(response.json())
ds_normalized = pd.DataFrame(norm_json)
ds_normalized.head(10)

ds_normalized.columns

from datetime import datetime

commit_times = ds_normalized['commit.author.date']
#commit_times = commit_times[1].astype('str')
commit_times = [datetime.strptime(commit_times[x], '%Y-%m-%dT%H:%M:%SZ') for x in range(len(commit_times))]
commit_times = pd.DataFrame(commit_times, columns = ['date'])
commit_times['date'] = pd.to_datetime(commit_times['date'])
commit_times

l7 = datetime(2019, 10, 28, 0, 0, 0)

last_7 = commit_times['date'] > l7
solution = np.count_nonzero(last_7)
solution
````
